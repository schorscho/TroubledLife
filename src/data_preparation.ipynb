```python
import os
from datetime import datetime
import data_preparation as sf


PROJECT_ROOT_DIR = "/home/ubuntu/TroubledLife"
#PROJECT_ROOT_DIR = "/Users/gopora/MyStuff/Dev/Workspaces/Sandbox/TroubledLife"
DATASETS_DIR = os.path.join(PROJECT_ROOT_DIR, "data")
TROUBLED_TRAINING_SET_DATA_FILE = "troubled_life_policy_train_data.csv"
TROUBLED_TEST_SET_DATA_FILE = "troubled_life_policy_test_data.csv"
PURE_TRAINING_SET_DATA_FILE = "pure_life_policy_train_data.csv"
PURE_TEST_SET_DATA_FILE = "pure_life_policy_test_data.csv"
```

```python
sf.generate_life_policy_data(
    no_of_policies=10000, runtime=5, trouble=True, 
    file_path=os.path.join(DATASETS_DIR, TROUBLED_TRAINING_SET_DATA_FILE))

sf.generate_life_policy_data(
    no_of_policies=2000, runtime=5, trouble=True, 
    file_path=os.path.join(DATASETS_DIR, TROUBLED_TEST_SET_DATA_FILE))

sf.generate_life_policy_data(
    no_of_policies=10000, runtime=5, trouble=False, 
    file_path=os.path.join(DATASETS_DIR, PURE_TRAINING_SET_DATA_FILE))

sf.generate_life_policy_data(
    no_of_policies=2000, runtime=5, trouble=False, 
    file_path=os.path.join(DATASETS_DIR, PURE_TEST_SET_DATA_FILE))
```

```python
troubled_policy_histories_train = \
    sf.load_life_policy_data(file_path=os.path.join(DATASETS_DIR, TROUBLED_TRAINING_SET_DATA_FILE))

troubled_policy_histories_test = \
    sf.load_life_policy_data(file_path=os.path.join(DATASETS_DIR, TROUBLED_TEST_SET_DATA_FILE))

pure_policy_histories_train = \
    sf.load_life_policy_data(file_path=os.path.join(DATASETS_DIR, PURE_TRAINING_SET_DATA_FILE))

pure_policy_histories_test = \
    sf.load_life_policy_data(file_path=os.path.join(DATASETS_DIR, PURE_TEST_SET_DATA_FILE))

troubled_policy_histories_length_train, troubled_max_policy_history_length_train = \
    sf.get_policy_history_lengths(policy_histories=troubled_policy_histories_train)

troubled_policy_histories_length_test, troubled_max_policy_history_length_test = \
    sf.get_policy_history_lengths(policy_histories=troubled_policy_histories_test)

pure_policy_histories_length_train, pure_max_policy_history_length_train = \
    sf.get_policy_history_lengths(policy_histories=pure_policy_histories_train)

pure_policy_histories_length_test, pure_max_policy_history_length_test = \
    sf.get_policy_history_lengths(policy_histories=pure_policy_histories_test)

max_policy_history_length = max(troubled_max_policy_history_length_train, troubled_max_policy_history_length_test,
                                pure_max_policy_history_length_train, pure_max_policy_history_length_test)
```

```python
# Pad the histories up to maximum length of both, train and test set
troubled_policy_histories_train = \
    sf.pad_life_policy_histories(policy_histories=troubled_policy_histories_train, 
                                          policy_histories_lengths=troubled_policy_histories_length_train,
                                          max_policy_history_length=max_policy_history_length)
    
troubled_policy_histories_test = \
    sf.pad_life_policy_histories(policy_histories=troubled_policy_histories_test,
                                          policy_histories_lengths=troubled_policy_histories_length_test,
                                          max_policy_history_length=max_policy_history_length)

pure_policy_histories_train = \
    sf.pad_life_policy_histories(policy_histories=pure_policy_histories_train, 
                                          policy_histories_lengths=pure_policy_histories_length_train,
                                          max_policy_history_length=max_policy_history_length)
    
pure_policy_histories_test = \
    sf.pad_life_policy_histories(policy_histories=pure_policy_histories_test,
                                          policy_histories_lengths=pure_policy_histories_length_test,
                                          max_policy_history_length=max_policy_history_length)
    
# Save padded data, since always generating and padding takes too long
troubled_policy_histories_train.to_csv(path_or_buf=os.path.join(DATASETS_DIR, TROUBLED_TRAINING_SET_DATA_FILE))
troubled_policy_histories_test.to_csv(path_or_buf=os.path.join(DATASETS_DIR, TROUBLED_TEST_SET_DATA_FILE))
pure_policy_histories_train.to_csv(path_or_buf=os.path.join(DATASETS_DIR, PURE_TRAINING_SET_DATA_FILE))
pure_policy_histories_test.to_csv(path_or_buf=os.path.join(DATASETS_DIR, PURE_TEST_SET_DATA_FILE))
```
